## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.show()
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
#sets fig equal to a figure
fig = plt.figure()

#sets ax equal to the axes
ax = fig.add_axes([0, 0, 1, 1]) 

# x and y in the color blue, labels the table axes x and y and title called title, and sets the x and y limits
ax.plot(x, y, 'b')
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_title('title')
ax.set_xlim([0,100])
ax.set_ylim([0,200])
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig,axes=plt.subplots(1,2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
#subplots with fig anad axes, with one row and 2 colums
fig,axes=plt.subplots(1,2)

#ax1 equal subplot and plots xy in blue, with width of 3 and dashed line
ax1=plt.subplot(1,2,1)
ax1.plot(x,y,'b',lw=3,ls='--')

#sets xlim and ylim for ax1
ax1.set_xlim([0,100])
ax1.set_ylim([0,200])

#ax2 equal subplot and plots xz in red, with width of 3
ax2=plt.subplot(1,2,2)
ax2.plot(x,z,'r',lw=3)

#ax2 sets xlim and ylim and does a tight layout
ax2.set_xlim([0,100]) 
ax2.set_ylim([0,10000])
plt.tight_layout()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
#subplots with fig anad axes, with one row and 2 colums, figsize is 7 by 2 inches
fig,axes=plt.subplots(1,2,figsize=(7,2))

#ax1 equal subplot and plots xy in blue, with width of 3 and dashed line, with plot number 1
ax1=plt.subplot(1,2,1)
ax1.plot(x,y,'b',lw=3,ls='--')

#sets xlim and ylim for ax1
ax1.set_xlim([0,100])
ax1.set_ylim([0,200])

#ax2 equal subplot and plots xz in blue, with width of 3 , with plot number 2
ax2=plt.subplot(1,2,2)
ax2.plot(x,z,'r',lw=3)

#sets xlim and ylim for ax2 and sets a tight layout
ax2.set_xlim([0,100]) 
ax2.set_ylim([0,10000])
plt.tight_layout()
```
