```python
#calculates the letter grade of the grade average of a class of students
"""
Author: Aryan Deorah
Source: Codeacdemy Module #6 Student Becomes the Teacher
"""
#Creates a dictionary for lloyd with keys for his name, homework, quizzes, and tests
lloyd = {
  "name": "Lloyd",
  "homework": [90.0, 97.0, 75.0, 92.0],
  "quizzes": [88.0, 40.0, 94.0],
  "tests": [75.0, 90.0]
}

#Creates a dictionary for alice with keys for his name, homework, quizzes, and tests
alice = {
  "name": "Alice",
  "homework": [100.0, 92.0, 98.0, 100.0],
  "quizzes": [82.0, 83.0, 91.0],
  "tests": [89.0, 97.0]
}

#Creates a dictionary for tyler with keys for his name, homework, quizzes, and tests
tyler = {
  "name": "Tyler",
  "homework": [0.0, 87.0, 75.0, 22.0],
  "quizzes": [0.0, 75.0, 78.0],
  "tests": [100.0, 100.0]
}

# Defines the average function that takes the argument, numbers. It sets the total equal to the sum of numbers and then to the decimal of the total
def average(numbers):
  total = sum(numbers)
  total = float(total)
  
  #returns total divided by the length of numbers
  return total/len(numbers)

#defines the get_average function with one argument, student
def get_average(student):
  
  #sets homework equal to the average of the key of 'homework' in student
  homework = average(student["homework"])
  
   #sets quizzes equal to the average of the key of 'quizzes' in student
  quizzes = average(student["quizzes"])
  
   #sets tests equal to the average of the key of 'tests' in student
  tests = average(student["tests"])
  
  # return the weighted average grade of the homework, quizzes and tests
  return 0.1 * homework + 0.3 * quizzes + 0.6 * tests

#defines the get_letter_grade function which takes one argument, score
def get_letter_grade(score):
  
  #checks if score is greater than or equal to 90 and returns the according letter grade
  if score >= 90:
    return "A"
  
   #checks if score is greater than or equal to 80 and returns the according letter grade
  elif score >=80:
    return "B"
  
   #checks if score is greater than or equal to 70 and returns the according letter grade
  elif score >=70:
    return "C"
  
   #checks if score is greater than or equal to 60 and returns the according letter grade
  elif score >=60:
    return "D"
  
   #if no other earlier statement is met, it returns the lowest possible letter grade
  else:
    return "F"
  
  #defines get_class_average with one argument, class_list and creates empty list called results
def get_class_average(class_list):
  results=[]
  
  #loops through the students and adds the weighted average of a student to the results list
  for student in class_list:
  	results.append(get_average(student))
    
  #returns the average of the results
  return average(results)

#defines list of students
students=[alice, lloyd, tyler]

#prints the class average and the letter grade of that average
print (get_class_average(students))
print (get_letter_grade(get_class_average(students)))
```
