```python
#creates a simplified version of the popular game Battleship
"""
Author: Aryan Deorah
Source: Codeacademy Module #7 Battleship
"""
#imports random integer function
from random import randint

#creates empty board list
board = []

#adds 5 '0's each in 5 items in the board list
for x in range(5):
  board.append(["O"] * 5)

  # defines print_board function with one argument, board. 
def print_board(board):
  
  #loops through each item in the board list and prints it in consecutive line
  for row in board:
    print (" ".join(row))
  
  #prints board onto scree
print_board(board)

#defines random_row function, taking board as an argument, and returns a random integer between 0 and one shorter than the length of the board list
def random_row(board):
  return randint(0, len(board) - 1)

#defines random_rcol function, taking board as an argument, and returns a random integer between 0 and one shorter than the length numbers in each item of the board list
def random_col(board):
  return randint(0, len(board[0]) - 1)

#sets variables ship_row and ship_col equal to the random_row and random_col functions, creating a coordinate for the hidden ship
ship_row = random_row(board)
ship_col = random_col(board)

#gives 4 turns to guess the location of the hidden ship
for turn in range(4):

  #sets guess_row and guess_col equal to the integers of raw inputs guessing the row and collumn of the hidden ship
  guess_row = int(input("Guess Row: "))
  guess_col = int(input("Guess Col: "))

  #if both the row and collumn guess were correct, it congradulates the play and breaks the loop
  if guess_row == ship_row and guess_col == ship_col:
    print ("Congratulations! You sunk my battleship!")
    break
  
  # if it was incorrect, it checks if the guess is not on the board and responds accordingly to your miss
  else:
    if (guess_row < 0 or guess_row > 4) or (guess_col < 0 or guess_col > 4):
      print ("Oops, that's not even in the ocean.")
    
    #if the spot had already been guessed, then it responds accordingly
    elif(board[guess_row][guess_col] == "X"):
      print ("You guessed that one already.")
    
    #if the spot was not guessed before and was not correct, it responds accordingly and stores that location on the map as an X
    else:
      print ("You missed my battleship!")
      board[guess_row][guess_col] = "X"
      
      #If it was the final turn, it says game over
      if turn==3:
        print ('Game Over')
  
  #it prints the next turn number and prints the board again
    print ("Turn", turn + 1)
    print_board(board)
```
